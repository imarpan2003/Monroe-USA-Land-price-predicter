```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#headers above
```

```python
data = pandas.read_csv('clean_data.csv')
data.describe()

#importing data file into Lab
```

```python
X = DataFrame(data, columns = ['area'])  #Feature
y = DataFrame(data, columns = ['price']) #Target

# Setting the value of target and feature
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.6)
plt.xlabel('Area')
plt.ylabel('Price for a Area')
plt.title('Hoe prices in Monroe Twp, NJ (USA)')
plt.show()

#Visualisation

```

 <h1> Regression </h1>

```python
regression = LinearRegression() #creating LinearRegression Object
```

```python
regression.fit(X, y)
print(regression.coef_)     #Slope Value
print(regression.intercept_) #Intercept Value
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(X, y, alpha = 0.6)

#Regression Line

plt.plot(X, regression.predict(X), color = 'green', linewidth = 3)


plt.xlabel('Area')
plt.ylabel('Price for a Area')
plt.title('Hoe prices in Monroe Twp, NJ (USA)')
plt.show()
```

```python
regression.score(X, y)
```
